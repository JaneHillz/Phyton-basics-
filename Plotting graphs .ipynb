```python
#plotting in matplotlib
#use scatterplot to see relationship between two varibales
#Use linegraph for continous variables 
#use boxplot to check outliers

X=dataset1['age']
y=dataset1['children']
plt.title("Age-Children_plot")
plt.xlabel('age')
plt.ylabel('children')
plt.plot(X,y)
plt.show()
```
